```python
#Abrimos el csv y lo guardamos en un dataframe llamado df_2
df_2 = pd.read_csv(directorio + "file_name", on_bad_lines='skip', encoding="utf-8")
df_2
```

```python
#a. Mostrar la cantidad y la media de películas producidas por año x plataforma. Crear un gráfico que muestre los resultados por plataforma.
# Guardamos en una lista las diferentes librerías. Dentro del bucle for agrupamos por año y contamos la cantidad de peliculas por plataforma.

plataformas = ["Netflix","Hulu","Prime Video","Disney+"]
x = pd.DataFrame()
y = pd.DataFrame()

for plataforma in plataformas:
    mascara = df_2[plataforma]==1
    df_filtrado = df_2[mascara].groupby(["Year"])
    cantidad_peliculas_año = df_filtrado.count()[plataforma]
    x[plataforma] = cantidad_peliculas_año
    
print(x)
x.plot()
```

```python
# b. Mostrar y crear un gráfico de las películas recomendadas por edad x plataforma
x = pd.DataFrame()

for plataforma in plataformas:
    mascara = df_2[plataforma]==1
    df_filtrado = df_2[mascara].groupby(["Age"])
    peliculas_recomendadas_edad = df_filtrado.count()[plataforma]
    x[plataforma] = peliculas_recomendadas_edad

print(x)

x.plot()
```

```python
# c. Mostrar las cinco películas mas vistas en general y por plataforma
x = pd.DataFrame()

for plataforma in plataformas:
    mascara = df_2[plataforma]==1
    df_fil = df_2[mascara].sort_values(by= "IMDb",ascending= False)
    peliculas_vistas_plataforma = df_fil["Title"].iloc[:5]
    peliculas_vistas_plataforma.index = np.arange(1,6)
    y = peliculas_vistas_plataforma
    
print(x)
```

```python
# d.Mostrar y crear un gráfico que muestre la cantidad de películas disponibles por plataforma
plataformas = ["Netflix","Hulu","Prime Video","Disney+"]
x = pd.DataFrame()


for plataforma in plataformas:
    mascara = df_2[plataforma]==1
    df_filtrado = df_2[mascara].count()[plataforma]
    x[plataforma] = pd.Series([df_filtrado])

x.loc[0].plot()



```

```python
# e. Mostrar en que plataformas está disponible una película (utilizar funciones)
# Creamos una función que recibe como parámetro el nombre de la pelicula, comprobamos para cada plataforma si para la película dada
# hay un 0 (no está en la plataforma) o un 1 (está en la plataforma). 

plataformas = ["Netflix","Hulu","Prime Video","Disney+"]
lista=[]

def busca_plataforma(pelicula):
    def aux(fila):
        for plataforma in plataformas:
            if fila[plataforma]==1:
                return plataforma

    df_filtrado = df_2.loc[df_2["Title"]==pelicula][plataformas]

    return df_filtrado.apply(aux,1)
    
# Mostramos unos ejemplos
print(busca_plataforma("Inception"))
print(busca_plataforma("The Poof Point"))
```

```python
#f. Muestre la media del IMDb rating por plataforma
plataformas = ["Netflix","Hulu","Prime Video",	"Disney+"]
x = pd.DataFrame()
lista=[]

for plataforma in plataformas:
    mascara = df_2[plataforma]==1
    df_filtrado = df_2[mascara]["IMDb"].mean()
    lista.append(df_filtrado)

medias = pd.Series(lista,index= plataformas)

print(medias)
```
